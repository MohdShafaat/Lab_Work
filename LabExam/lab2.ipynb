```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("DataCovid.csv")
```

```python
df
```

```python
newDF = df.drop(df.columns[0], axis = 1)
mat = newDF.to_numpy()
print(mat)
```

```python
row = mat.shape[0]-1 ## This R does not consist of the last row to be predicted
col = mat.shape[1]
```

```python
countYes=0
for i in range(row):
    if mat[i][col-1] == "Yes":
        countYes+=1
countNo=row-countYes
P = countYes/row ## P for P(E=Yes)
NP = countNo/row ## NP for P(E=No)
```

```python
Pyes=1
Pno=1
for c in range(col-1):
    instanceVal=mat[row][c]
    countInstanceYes=0
    countInstanceNo=0
    for r in range(row):
        if mat[r][c]==instanceVal:
            if mat[r][col-1]=="Yes":
                countInstanceYes+=1
            else:
                countInstanceNo+=1
    Pyes=Pyes*(countInstanceYes/countYes)
    Pno=Pno*(countInstanceNo/countNo)
Pyes=Pyes*P
Pno=Pno*NP
print(f"Yes = {Pyes}")
print(f"No = {Pno}")
```

```python
if Pyes>Pno:
    df.iloc[row, col] = "Yes"
else:
    df.iloc[row, col] = "No"
```
